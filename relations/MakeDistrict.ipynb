```python
import pandas as pd
```

```python
df = pd.read_csv('Education.csv')
```

```python
df = df[['districtID']]
```

```python
import re 
state = []
districtNum = []
for idx, row in df.iterrows():
    res = [re.findall(r'(\w+?)(\d+)', row['districtID'])[0]]
    st = res[0][0]
    if '_' in st:
        lst = st.split('_')
        temp = ''
        for i in range(len(lst)):
            temp += lst[i].capitalize()
            if i == 0:
                temp += ' '
        state.append(temp)
    else:
        state.append(st.capitalize())
    districtNum.append(res[0][1])
```

```python
df['state'] = state
df['congressionalDistrictNum'] = districtNum
```

```python
df = df.set_index('districtID')
```

```python
df.to_csv('Districts.csv')
```
